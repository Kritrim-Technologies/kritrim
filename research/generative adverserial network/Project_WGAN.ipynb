```python
class ClipConstraint(Constraint):
	# set clip value when initialized
	def __init__(self, clip_value):
		self.clip_value = clip_value
 
	# clip model weights to hypercube
	def __call__(self, weights):
		return backend.clip(weights, -self.clip_value, self.clip_value)
 
	# get the config
	def get_config(self):
		return {'clip_value': self.clip_value}
```

```python
def wasserstein_loss(y_true, y_pred):
  return tf.reduce_mean(y_true * y_pred)
```

```python
def generator():
    gen_input = Input(shape = (None, 100))
    X = LSTM(256, return_sequences = True)(gen_input)
    X = BatchNormalization(momentum = 0.8)(X)
    X = LeakyReLU(alpha = 0.2)(X)
    X = LSTM(128, return_sequences = True)(X)
    X = BatchNormalization(momentum = 0.8)(X)
    X = LeakyReLU(alpha = 0.2)(X)
    X = Dense(64)(X)
    X = LeakyReLU(alpha = 0.2)(X)
    X = BatchNormalization(momentum = 0.8)(X)
    X = Dropout(0.5)(X)
    out = Dense(2, activation = "tanh")(X)
    model = Model(inputs = gen_input, outputs = out)
    return model
```

```python
def discriminator():
    disc_input = Input(shape = (100, 2))
    const = ClipConstraint(0.01)
    X = LSTM(128, return_sequences = True, kernel_constraint = const)(disc_input)
    X = LeakyReLU(alpha = 0.2)(X)
    X = LSTM(64, return_sequences = True, kernel_constraint = const)(X)
    X = LeakyReLU(alpha = 0.2)(X)
    X = Flatten()(X)
    X = Dense(128, kernel_constraint = const)(X)
    X = LeakyReLU(alpha = 0.2)(X)
    out = Dense(1)(X)
    model = Model(inputs = disc_input, outputs = out)
    return model
```

```python
def disc_with_gen(generator, discriminator):
    inputs = Input(shape = (None, 100))
    generated = generator(inputs)
    outputs = discriminator(generated)
    model = Model(inputs = inputs, outputs = outputs)
    return model
```

```python
from tensorflow.keras.optimizers import RMSprop
optimizer = RMSprop(lr = 0.00005)
d = discriminator()
d.compile(loss = wasserstein_loss, optimizer=optimizer, metrics=['accuracy'])
g = generator()
```

```python
d.trainable = False
combined = disc_with_gen(g, d)
combined.compile(loss = wasserstein_loss, optimizer=optimizer, metrics=['accuracy'])
```

```python
def train(epochs, num_of_batches, batch_size = batch_size, n_critic = 5):
    for epoch in range(epochs):
      a = 0
      b = batch_size
      for num in range(num_of_batches):
        idx = list(range(a, b, 1))
        real_data = X[idx]

        noise = np.random.normal(0, 1, (batch_size, 100, 100))

        real_loss = []
        fake_loss = []

        for _ in range(n_critic):
          generated = g.predict(noise)
          d_loss_real = d.train_on_batch(real_data, -(np.ones((batch_size, 1))))
          real_loss.append(d_loss_real)
          #print("d_loss_real: ", d_loss_real)
          d_loss_fake = d.train_on_batch(generated, np.ones((batch_size, 1)))
          fake_loss.append(d_loss_fake)
          #print("d_loss_fake: ",d_loss_fake)

        disc_loss_real.append(np.mean(real_loss))
        print("%d d_loss_real: " %epoch, np.mean(real_loss))
        disc_loss_fake.append(np.mean(fake_loss))
        print("%d d_loss_fake: " %epoch, np.mean(fake_loss))

        noise = np.random.normal(0, 1, (batch_size, 100, 100))
        g_loss = combined.train_on_batch(noise, -(np.ones((batch_size, 1))))
        print("%d g_loss: " %epoch, g_loss)

        print("\n")

        gen_loss.append(g_loss[0])

        a = a + batch_size
        b = b + batch_size
            
      if epoch % 1 == 0:
        disc_loss_real_epoch.append(np.mean(real_loss))
        disc_loss_fake_epoch.append(np.mean(fake_loss))
        gen_loss_epoch.append(g_loss[0])
        #save_weights()
        #save_models()
        #save_losses()
        checkpoint = tf.train.Checkpoint(g)
        checkpoint.save('/content/drive/MyDrive/Major Project/WGAN/checkpoints/training_checkpoints')
            
    plt.plot(disc_loss, c='red')
    plt.plot(gen_loss, c='blue')
    plt.title("GAN Loss per batch")
    plt.legend(['Discriminator', 'Generator'])
    plt.xlabel('Batch')
    plt.ylabel('Loss')
    plt.savefig('/content/drive/MyDrive/Major Project/output/WGAN_Loss_per_Batch_final.png', transparent=False)
    plt.close()

    plt.plot(disc_loss_epoch, c='red')
    plt.plot(gen_loss_epoch, c='blue')
    plt.title("GAN Loss per Epoch")
    plt.legend(['Discriminator', 'Generator'])
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.savefig('/content/drive/MyDrive/Major Project/output/WGAN_Loss_per_Epoch_final.png', transparent=False)
    plt.close()
```
